```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# City data: (name, latitude, longitude)
cities = [
    ("Georgia", 31.7642, -82.3519),
    ("New York", 44.7289, -73.5383),
    ("WA", 48.7519, -122.4787),
    ("Alabama", 34.0817, -86.5917),
    ("Texas", 25.9969, -97.1632),
    ("Illinois", 41.8584, -87.8361),
    ("Montana", 45.9412, -112.5868),
    ("North Dakota", 47.0294, -97.2035),
    ("Oregon", 43.3726, -124.2365),
    ("Idaho", 48.4218, -116.7359),
    ("MT", 48.1718, -111.935),
    ("Colorado", 39.2423, -104.528),
    ("Wyoming", 41.2683, -110.9632),
    ("Florida", 26.1224, -80.1373),
    ("Maryland", 39.4142, -77.4105),
    ("South Carolina", 35.0729, -81.6487),
]

# Set up the map with more area around the USA
plt.figure(figsize=(18, 15))
m = Basemap(
    projection='merc',
    llcrnrlat=20, urcrnrlat=55,  # Adjusted latitude range
    llcrnrlon=-140, urcrnrlon=-50,  # Adjusted longitude range
    resolution='l'
)

# Draw coastlines, states, and countries
m.drawcoastlines()
m.drawstates()
m.drawcountries()

# Plot cities with adjusted styles
for city, lat, lon in cities:
    x, y = m(lon, lat)
    m.plot(x, y, 'ro', markersize=16, alpha=1)  # Bigger red dots
    plt.text(x, y, city, color='#278704', fontsize=14, fontweight='bold', alpha=0.8)

# Drawing lines between cities
# WA to oregon
lat_i, lon_i = cities[2][1], cities[2][2]
lat_j, lon_j = cities[8][1], cities[8][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# WA to idaho
lat_i, lon_i = cities[2][1], cities[2][2]
lat_j, lon_j = cities[9][1], cities[9][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# montana to mt
lat_i, lon_i = cities[9][1], cities[9][2]
lat_j, lon_j = cities[10][1], cities[10][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# Montana to idaho
lat_i, lon_i = cities[9][1], cities[9][2]
lat_j, lon_j = cities[6][1], cities[6][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# MT to Nort dakota.
lat_i, lon_i = cities[10][1], cities[10][2]
lat_j, lon_j = cities[7][1], cities[7][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# North dakota to NYork.
lat_i, lon_i = cities[7][1], cities[7][2]
lat_j, lon_j = cities[1][1], cities[1][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# Nyork to maryland.
lat_i, lon_i = cities[1][1], cities[1][2]
lat_j, lon_j = cities[14][1], cities[14][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# Maryland to south caroline.
lat_i, lon_i = cities[14][1], cities[14][2]
lat_j, lon_j = cities[15][1], cities[15][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# south karolina to georgia.
lat_i, lon_i = cities[15][1], cities[15][2]
lat_j, lon_j = cities[0][1], cities[0][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# georgia to florida.
lat_i, lon_i = cities[0][1], cities[0][2]
lat_j, lon_j = cities[13][1], cities[13][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

#gerogia to alabama.
lat_i, lon_i = cities[0][1], cities[0][2]
lat_j, lon_j = cities[3][1], cities[3][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# alabama to texas.
lat_i, lon_i = cities[3][1], cities[3][2]
lat_j, lon_j = cities[4][1], cities[4][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# MT to wyoming.
lat_i, lon_i = cities[10][1], cities[10][2]
lat_j, lon_j = cities[12][1], cities[12][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# illionois to north dakota.
lat_i, lon_i = cities[5][1], cities[5][2]
lat_j, lon_j = cities[7][1], cities[7][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# texas to colorodo.
lat_i, lon_i = cities[4][1], cities[4][2]
lat_j, lon_j = cities[11][1], cities[11][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# texas to oregon.
lat_i, lon_i = cities[4][1], cities[4][2]
lat_j, lon_j = cities[8][1], cities[8][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# montana to oregon.
lat_i, lon_i = cities[6][1], cities[6][2]
lat_j, lon_j = cities[8][1], cities[8][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# wyoming to northdakota.
lat_i, lon_i = cities[12][1], cities[12][2]
lat_j, lon_j = cities[7][1], cities[7][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# ml to illionoi
lat_i, lon_i = cities[14][1], cities[14][2]
lat_j, lon_j = cities[5][1], cities[5][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# albama to colorodao
lat_i, lon_i = cities[3][1], cities[3][2]
lat_j, lon_j = cities[11][1], cities[11][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# col to wyoming
lat_i, lon_i = cities[11][1], cities[11][2]
lat_j, lon_j = cities[12][1], cities[12][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# alb to florida
lat_i, lon_i = cities[3][1], cities[3][2]
lat_j, lon_j = cities[13][1], cities[13][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# south carolinie to illionoi
lat_i, lon_i = cities[15][1], cities[15][2]
lat_j, lon_j = cities[5][1], cities[5][2]
x_i, y_i = m(lon_i, lat_i)
x_j, y_j = m(lon_j, lat_j)
plt.plot([x_i, x_j], [y_i, y_j], color='red', linestyle='-', linewidth=4, alpha=0.6)

# Remove axes, labels, and title
plt.axis('off')

# Save the graph as a PDF
plt.savefig('cities_map.pdf', bbox_inches='tight', pad_inches=0, format='pdf')
plt.show()
```
