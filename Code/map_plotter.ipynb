```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap

# City data: (name, latitude, longitude)
cities = [
    # ("Alaska", 61.2173, -149.8631),
    ("Georgia", 31.7642, -82.3519),
    ("New York", 44.7289, -73.5383),
    ("WA", 48.7519, -122.4787),
    ("Alabama", 34.0817, -86.5917),
    ("Texas", 25.9969, -97.1632),
    ("Illinois", 41.8584, -87.8361),
    ("Montana", 45.9412, -112.5868),
    ("North Dakota", 47.0294, -97.2035),
    ("Oregon", 43.3726, -124.2365),
    ("Idaho", 48.4218, -116.7359),
    ("MT", 48.1718, -111.935),
    ("Texas", 35.8614, -101.9634),
    ("Colorado", 39.2423, -104.528),
    ("Wyoming", 41.2683, -110.9632),
    # ("Alaska", 64.8378, -147.7164),
    ("Florida", 26.1224, -80.1373),
    ("Maryland", 39.4142, -77.4105),
    ("South Carolina", 35.0729, -81.6487),
]

# Set up the map with more area around the USA
plt.figure(figsize=(18, 15))
m = Basemap(
    projection='merc',
    llcrnrlat=20, urcrnrlat=55,  # Adjusted latitude range
    llcrnrlon=-140, urcrnrlon=-50,  # Adjusted longitude range
    resolution='l'
)

# Draw coastlines, states, and countries
m.drawcoastlines()
m.drawstates()
m.drawcountries()

# Plot cities with adjusted styles
for city, lat, lon in cities:
    x, y = m(lon, lat)
    m.plot(x, y, 'ro', markersize=16, alpha=1)  # Bigger red dots
    plt.text(x, y, city, color='#278704', fontsize=14, fontweight='bold', alpha=0.8)

# Connect cities with lines
for i in range(len(cities)):
    for j in range(i + 1, len(cities)):
        lat_i, lon_i = cities[i][1], cities[i][2]
        lat_j, lon_j = cities[j][1], cities[j][2]
        x_i, y_i = m(lon_i, lat_i)
        x_j, y_j = m(lon_j, lat_j)
        plt.plot([x_i, x_j], [y_i, y_j], color='blue', linestyle='-', linewidth=1, alpha=0.6)

# Remove axes, labels, and title
plt.axis('off')

# Save the graph as a PDF
plt.savefig('cities_map.pdf', bbox_inches='tight', pad_inches=0, format='pdf')
plt.show()
```
